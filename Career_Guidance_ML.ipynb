```python
import pandas as pd
url='https://raw.githubusercontent.com/thecodebloodedguy/career-guidance-ml-model/main/dataset-github.csv'
df=pd.read_csv(url)
```

```python
df


```

```python
for col in df.select_dtypes(include=['object']).columns:
    df[col] = df[col].astype('category')
    df[col] = df[col].cat.codes
df.dtypes
```

```python
df.dropna()
df.isna().sum()
```

```python
df
```

```python
X = df.drop(['Suggested Job Role'], axis=1)
Y=df['Suggested Job Role']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
lnr = LogisticRegression()
lnr.fit(X_train, Y_train)
```

```python
Y_lnr_train_pred = lnr.predict(X_train)
Y_lnr_test_pred = lnr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
lnr_train_mse = mean_squared_error(Y_train, Y_lnr_train_pred)
lnr_train_r2 = r2_score(Y_train, Y_lnr_train_pred)
lnr_test_mse = mean_squared_error(Y_test, Y_lnr_test_pred)
lnr_test_r2 = r2_score(Y_test, Y_lnr_test_pred)
```

```python
print(lnr_train_mse)
```

```python
lnr_results = pd.DataFrame(['Logistic Regression',lnr_train_mse, lnr_train_r2, lnr_test_mse, lnr_test_r2]).transpose()
lnr_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
```

```python
lnr_results
```

```python
from sklearn.tree import DecisionTreeClassifier
rf = DecisionTreeClassifier()
rf.fit(X_train, Y_train)
```

```python
Y_rf_train_pred = rf.predict(X_train)
Y_rf_test_pred = rf.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rf_train_mse = mean_squared_error(Y_train, Y_rf_train_pred)
rf_train_r2 = r2_score(Y_train, Y_rf_train_pred)
rf_test_mse = mean_squared_error(Y_test, Y_rf_test_pred)
rf_test_r2 = r2_score(Y_test, Y_rf_test_pred)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_m = RandomForestClassifier(max_features=2, random_state=42, oob_score=True)
rf_m.fit(X_train, Y_train)
```

```python
Y_rfm_train_pred = rf_m.predict(X_train)
Y_rfm_test_pred = rf_m.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
rfm_train_mse = mean_squared_error(Y_train, Y_rfm_train_pred)
rfm_train_r2 = r2_score(Y_train, Y_rfm_train_pred)
rfm_test_mse = mean_squared_error(Y_test, Y_rfm_test_pred)
rfm_test_r2 = r2_score(Y_test, Y_rfm_test_pred)
```

```python
rf_results = pd.DataFrame(['Descision Tree Classifier',rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
```

```python
rf_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
rf_results
```

```python
rfm_results = pd.DataFrame(['Random forest Classifier',rfm_train_mse, rfm_train_r2, rfm_test_mse, rfm_test_r2]).transpose()
```

```python
rfm_results.columns = ['Method','Training MSE','Training R2','Test MSE','Test R2']
rfm_results
```

```python
res_list=[rfm_results,rf_results,lnr_results]
results=pd.concat(res_list)
```

```python
results
```
